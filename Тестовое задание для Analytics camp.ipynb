1. Вася выписал на доску натуральные числа, все они различны, при том часть из них делятся
на 2 но не делятся на 3, а остальные делятся на 3, но не делятся на 2. Петя посмотрел на
выписанные на доске числа и заметил, что можно 25 способами выбрать три числа так, чтобы
среди них было как минимум одно четное и хотя бы одно число, делящееся на 3.
Сколько чисел написал Вася?

Решение:

     Пусть количество чисел, которые написал Вася и делятся на 2, равно a, а количество чисел, которые написал Вася и делятся на 3, равно b. Также пусть количество чисел, которые написал Вася и не делятся ни на 2, ни на 3, равно c. Тогда общее количество чисел, которые написал Вася, равно:
    a + b + c.

    Мы знаем, что a + b = 25, так как мы должны выбрать три числа из всех чисел, и хотя бы одно из них должно быть четным, а хотя бы одно должно делиться на 3.

    Также мы знаем, что c = 0, так как все числа на доске делятся либо на 2, либо на 3.

    Таким образом, мы можем записать a = 25 - b. Теперь нам нужно найти b.

    Заметим, что все числа, которые делятся на 2, но не делятся на 3, можно записать в виде 2k, где k - некоторое натуральное число. Аналогично, все числа, которые делятся на 3, но не делятся на 2, можно записать в виде 3m, где m - некоторое натуральное число.

    Таким образом, мы можем записать:

    b = 3m
    a = 25 - b = 25 - 3m

    Теперь нам нужно найти все возможные значения m, для которых a и b являются натуральными числами.

    Мы знаем, что a > 0 и b > 0, так как каждое из них представляет количество чисел, которые написал Вася и делятся на 2, и на 3 соответственно. Поэтому мы можем записать:

    25 - 3m > 0
    3m < 25

    Решая неравенство, получаем:

    0 < m < 8.33

    Таким образом, m может принимать значения 1, 2, 3, 4, 5, 6, 7. Подставляя каждое значение m, мы можем найти соответствующие значения a и b:

    m = 1: a = 22, b = 3
    m = 2: a = 19, b = 6
    m = 3: a = 16, b = 9
    m = 4: a = 13, b = 12
    m = 5: a = 10, b = 15
    m = 6: a = 7, b = 18
    m = 7: a = 4, b = 21

    Мы видим, что только для m = 4 и m = 5 выполняется условие, что a и b являются натуральными числами. Поэтому число написанных Васей чисел равно a + b + c = 25, так как все числа на доске делятся на 2 или на 3.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import warnings
warnings.filterwarnings("ignore")
```

```python
df_model = data[data['medium'] == 'cpc'][
    ['date', 'pageviews', 'visits', 'productClick', 'addToCart', 'checkout','transactions', 'revenue']]
df_model['date'] = pd.to_datetime(df_model['date'], format='%Y-%m')
df_model
```

```python
def generate_degrees(source_data: list, degree: int):
    """Функция, которая принимает на вход одномерный массив, а возвращает n-мерный
    Для каждой степени от 1 до  degree возводим x в эту степень
    """
    return np.array([
          source_data**n for n in range(1, degree + 1)  
    ]).T
```

```python
degree = 8

X = generate_degrees(data['x_train'], degree)
y = data.y_train.values

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=10)
model = Ridge(alpha=0).fit(X_train, y_train)
y_pred = model.predict(X_valid)
y_pred_train = model.predict(X_train)
```

```python
X = df_model[['pageviews', 'visits', 'productClick','addToCart', 'checkout', 'transactions']] #фичи
y = df_model['revenue'] # целевой признак

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# обучение линейной регрессии
lr = LinearRegression()
lr = Lasso(alpha=0.01).fit(X_train, y_train)
pred_train_lr = lr.predict(X_train)
pred_valid_lr = lr.predict(X_valid) 

print("Качество на обучении: %.3f" % r2_score(y_train, pred_train_lr))
print("Качество на валидации: %.3f" % r2_score(y_valid, pred_valid_lr))
```

```python
# обучение случайного леса
fr = RandomForestRegressor(n_estimators=100)
fr = fr.fit(X_train, y_train)
pred_train_fr = fr.predict(X_train)
pred_valid_fr = fr.predict(X_valid) 

print("Качество на обучении: %.3f" % r2_score(y_train, pred_train_fr))
print("Качество на валидации: %.3f" % r2_score(y_valid, pred_valid_fr))
```

```python
# обучение градиентного бустинга
gb = GradientBoostingRegressor(n_estimators=100)
gb = gb.fit(X_train, y_train)
pred_train_gb = gb.predict(X_train)
pred_valid_gb = gb.predict(X_valid) 

print("Качество на обучении: %.3f" % r2_score(y_train, pred_train_gb))
print("Качество на валидации: %.3f" % r2_score(y_valid, pred_valid_gb))
```

Исходя из метрики качества выбираю модель Random Forest Regressor

    console.cloud.google.com/bigquery?p=bigquery-public-data&d=baseball. Чтобы работать в
    интерфейсе BigQuery, в Enter project name добавь bigquery-public-data.
    В качестве ответа на данный вопрос необходимо предоставить SQL запрос, который выводит
    необходимые данные, а также сами эти данные, отсортированные по названию команды в
    порядке возрастания.

      SQL код:

    SELECT 
      teams.name AS team_name, 
      players.name_first AS first_name, 
      players.name_last AS last_name, 
      MAX(pitches.release_speed) AS max_release_speed 
    FROM 
      `bigquery-public-data.baseball.games_wide` AS games
    INNER JOIN 
      `bigquery-public-data.baseball.pitches` AS pitches 
    ON 
      games.game_id = pitches.game_id 
    INNER JOIN 
      `bigquery-public-data.baseball.players` AS players 
    ON 
      players.player_id = pitches.pitcher_id 
    INNER JOIN 
      `bigquery-public-data.baseball.teams` AS teams 
    ON 
      teams.team_id = pitches.pitching_team_id 
    WHERE 
      pitches.release_speed IS NOT NULL 
    GROUP BY 
      team_name 
    ORDER BY 
      team_name ASC
